# Split Training and Testing Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_np, 
                                                    y, 
                                                    test_size = 0.2, 
                                                    random_state = 42 )
```

```python
print(X_train.shape, X_test.shape)
print(y_train.shape, y_test.shape)
```

```python
X_train_cs = X_train / 255
X_test_cs = X_test / 255
```

# K - Nearest Neighbors - Understanding Indices, Distances

```python

from sklearn.neighbors import NearestNeighbors
example_amnt = len(X_train)
k = int(math.sqrt(example_amnt))          

knn = NearestNeighbors(n_neighbors=k).fit(X_train_cs)    
distances, indices = knn.kneighbors([X_test_cs[0]], n_neighbors = k)

print("indices:")
print(indices)
print()
print("distances:")
print(distances)
```

```python
print("indices:")
print(indices)
print()
print("indices shape:")
print(indices.shape)     
print()
print("indices shape:")
print(indices.squeeze().shape)  

print("\n- - - - - \n")


print("indices:")
print(indices)
print()
print("indices shape:")
print(indices.squeeze()) 
print()

print("\n- - - - - \n")


print("indices reshape:")
print(indices.reshape(k,)) 
print()
```

```python
print("indices:")
print(indices)
print()
print("distances:")
print(distances)
print()
print("Training Values:")
#print(y_train[indices.squeeze()])   
print(y[indices.squeeze()])   
```
